```python
from pyiceberg.catalog import load_catalog
import polars as pl
from pathlib import Path
warehouse_path = Path("iceberg_catalog").absolute()
catalog = load_catalog(
    name="local",
    **{
        "type": "sql",
        "uri": "sqlite:///iceberg_catalog/catalog.db",
        "warehouse": f"file://{warehouse_path}"
    }
)
#catalog = load_catalog("local")
product_table = catalog.load_table('Sales.Product')
```

```python
print(catalog.properties)
```

```python
scanned_products = pl.scan_iceberg(product_table)
```

```python
df_products = scanned_products.select("Name")
```

```python
df_selected = df_products.collect()
```
